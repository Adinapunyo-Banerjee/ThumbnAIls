```python
import requests
from bs4 import BeautifulSoup

# Scraping from socialblade.com
url = 'https://socialblade.com/youtube/top/category/travel'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'}
```

```python
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
usernames = []
```

```python
for row in soup.select('a[href*="/youtube/c/"]'):
    usernames.append(row.text)
    # print(row)

# for row in soup.select('body > div > div > div > a'):
#     print(row)

print(usernames)


def write_list_to_txt_file(my_list, file_path):
    # Open the file in write mode
    with open(file_path, 'a', encoding='utf-8') as file:
        # Loop through the list and write each item on a separate line
        for item in my_list:
            file.write(str(item) + '\n')
c = int(input("Do you want to write this to file? (0/1)"))

# Example usage:
if c==1:
  file_path = '/content/drive/MyDrive/Colab Notebooks/Youtube Project/channel_names.txt'
  write_list_to_txt_file(usernames, file_path)
  print("Done!")
else:
  print("Cancelled!")
```
